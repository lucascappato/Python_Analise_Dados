```python
import pandas as pd

vendas = [
    {"produto": "Notebook", "categoria": "Eletrônicos", "valor": 3500, "cliente": "Lucas"},
    {"produto": "Mouse", "categoria": "Eletrônicos", "valor": 150, "cliente": "Pedro"},
    {"produto": "Teclado", "categoria": "Eletrônicos", "valor": 300, "cliente": "Ana"},
    {"produto": "Cadeira Gamer", "categoria": "Móveis", "valor": 1200, "cliente": "Lucas"},
    {"produto": "Monitor", "categoria": "Eletrônicos", "valor": 900, "cliente": "Marcelo"},
    {"produto": "Mesa Escritório", "categoria": "Móveis", "valor": 700, "cliente": "Pedro"},
    {"produto": "Headset", "categoria": "Eletrônicos", "valor": 400, "cliente": "Ana"},
    {"produto": "Notebook", "categoria": "Eletrônicos", "valor": 4200, "cliente": "Marcelo"}
]

df = pd.DataFrame(vendas)

df
```

```python
df["valor"].mean()
```

```python
df["tipo_venda"] = "Normal"

df.loc[df["valor"] >= 2000, "tipo_venda"] = "Alta"

df
```

```python
df["tipo_venda"].value_counts()
```
